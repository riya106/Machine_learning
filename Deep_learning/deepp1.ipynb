```python
X_tarin=X_train /255
X_test=X_test/255
```

```python
#there is a function in numpy as reshape so as we falttening the array to one dimesion so inseted of 28 two times we will convert taht into 28*28 as one 
X_train_flatten=X_train.reshape(len(X_train),28*28)
X_test_flatten=X_test.reshape(len(X_test),28*28)
```

```python
X_train_flatten.shape
```

```python
X_test_flatten.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flatten, y_train, epochs=5)
```

```python
model.evaluate(X_test_flatten,y_test)
```

```python
model.predict(X_test_flatten)
```

```python
plt.matshow(X_test[0])
```

```python
y_predicted=model.predict(X_test_flatten)
```

```python
y_predicted[0]
```

```python
#this is the function whch predict the max value from the  these array and then return the index
np.argmax(y_predicted[0])
```

```python
y_predicted_label=[np.argmax(i) for i in y_predicted]
y_predicted_label[:5]
```

```python
y_test[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_label)
cm
```

```python
import seaborn as sn
plt.figure(figsize =(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.ylabel('Truth')
```

```python
#here we are adding the new layer
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),#here we add one hidden layer 
    keras.layers.Dense(10, activation='sigmoid')
    

])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flatten, y_train, epochs=5)
```

```python
model.evaluate(X_test_flatten,y_test)
```

```python

#here we are trying to deal with the flatten we are actually having a inbuilt function of the flatten so we need to not to flatten it 
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),#here we add one hidden layer 
    keras.layers.Dense(10, activation='sigmoid')
    

])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train, y_train, epochs=5)
```

```python

#here we will chnge the value of different dofferent facyors and see what are the vhages in the accuracy
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='sgd',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train, y_train, epochs=20)
```

```python
model.evaluate(X_test,y_test)
```
